```python
import numpy as np
import pandas as pd
from pandasql import sqldf
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_excel('/content/sample_data/Interview Project Data.xlsx', sheet_name='Ticket Scan Data')
df['Dates'] = pd.to_datetime(df['event_datetime']).dt.date
df['Time'] = pd.to_datetime(df['event_datetime']).dt.time
df['Year'] = pd.DatetimeIndex(df['Dates']).year
```

```python
example = sqldf("SELECT year, event_name, event_datetime, scan_datetime, Dates, Time, count(event_name) as Attendance FROM df GROUP BY Dates")
example.head()
```

```python
example.describe()
example.set_index('Year',inplace=True)

from pylab import rcParams
rcParams['figure.figsize'] = 15, 7
example.plot()
```

```python
from statsmodels.tsa.stattools import adfuller
test_result=adfuller(example['Attendance'])

def AdfullerTest(attendance):
    result=adfuller(attendance)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )

result=adfuller(example['Attendance'])
if result[1] <= 0.05:
    print("Data is stationary")
else:
    print("Data is non-stationary ")

AdfullerTest(example['Attendance'])
#Here P-value is 0.36 which is greater than 0.05, which means data is accepting the null hypothesis, which means data is non-stationary.
#As mentioned before, a time series must be stationary for it to be used to predict well founded values
#The null hypothesis is that the term is not significantly different from 0, which indicates that no association exists between the term and the response.
```
